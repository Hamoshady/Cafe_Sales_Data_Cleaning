```python
# Edit columns names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
print(f'new columns is: {df.columns.to_list()}')
```

```python
# Convert ERROR and UNKNOWN to NaN
cleanup_map = {
    'ERROR' : np.nan,
    'UNKNOWN' : np.nan,
}

df = df.replace(cleanup_map)
```

```python
# Convert the datatype for this columns
df['quantity'] = pd.to_numeric(df['quantity'], errors = 'coerce')
df['price_per_unit'] = pd.to_numeric(df['price_per_unit'], errors = 'coerce')
df['total_spent'] = pd.to_numeric(df['total_spent'], errors = 'coerce')
df['transaction_date'] = pd.to_datetime(df['transaction_date'], errors = 'coerce')
```

```python
mask = (df['quantity'] * df['price_per_unit'] != df['total_spent'])
print(df.loc[mask, 'transaction_id'])
```

```python
price_map = {
    'Cake' : 3.0,
    'Coffee' : 2.0,
    'Cookie' : 1.0,
    'Juice' : 3.0,
    'Salad' : 5.0,
    'Sandwich' : 4.0,
    'Smoothie' : 4.0,
    'Tea' : 1.5
}
df['price_per_unit'] = df['price_per_unit'].fillna(df['item'].map(price_map))
```

```python
price_map_for_item = {
    'Coffee' : 2.0,
    'Cookie' : 1.0,
    'Salad' : 5.0,
    'Tea' : 1.5
}
item_map = {v: k for k,v in price_map_for_item.items()}
df['item'] = df['item'].fillna(df['price_per_unit'].map(item_map))
```

```python
df.loc[df['quantity'].isna(), 'quantity'] = (df['total_spent'] / df['price_per_unit'])
df.loc[df['price_per_unit'].isna(), 'price_per_unit'] = (df['total_spent'] / df['quantity'])
df.loc[df['total_spent'].isna(), 'total_spent'] = (df['quantity'] * df['price_per_unit'])
```
